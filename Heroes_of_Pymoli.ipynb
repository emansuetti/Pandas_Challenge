```python
import pandas as pd
import numpy as np

file_to_load = "purchase_data.csv"

purchase_data = pd.read_csv(file_to_load)

purchase_data.head(20)
```

```python
#Player Count
purchase_data_df = pd.DataFrame(purchase_data)

player_count = purchase_data_df["SN"].count()

player_count_df = pd.DataFrame([{"Total Players": player_count}])
player_count_df
```

```python
#Unique Item Count
unique_item_count = len(purchase_data_df["Item ID"].unique())
unique_item_count
```

```python
#Total number of purchases
num_purchases = purchase_data_df["Purchase ID"].count()
num_purchases
```

```python
#Total Revenue
total_price = purchase_data_df["Price"].sum()
total_price
```

```python
#Average Price
average_price = purchase_data_df["Price"].mean()
average_price
```

```python
#Analysis and table
purchase_analysis_df = pd.DataFrame([{"Number of Unique items": unique_item_count, 
                                     "Average Price": average_price,
                                     "Number of Purchases": num_purchases,
                                     "Total Revenue": total_price}])
purchase_analysis_df["Average Price"] = purchase_analysis_df["Average Price"].map("${:.2f}".format)
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map("${:,.2f}".format)

purchase_analysis_df
```

```python
#Gender Demographics
gender_df = pd.DataFrame(purchase_data_df["Gender"].value_counts())
gender_df
```

```python
#Percentage of male, female, and other players
percentage_df = (purchase_data_df["Gender"].value_counts()/player_count)*100
percentage_df                   
```

```python
gender_analysis_df = pd.DataFrame(gender_df)
#gender_analysis_df

gender_analysis_df["Total Count"] = gender_df
gender_analysis_df["Percentage of Players"] = percentage_df
gender_analysis_df["Percentage of Players"] = gender_analysis_df["Percentage of Players"].map("{:.2f}%".format)
gender_analysis_df.drop('Gender', inplace=True, axis=1)
gender_analysis_df
```

```python
gender_group = purchase_data_df.groupby(["Gender"])
#gender_group.count().head()

male_total = (purchase_data_df['Gender'] =='Male').sum()
#male_total

female_total = (purchase_data_df['Gender'] =='Female').sum()
#female_total

other_total = (purchase_data_df['Gender'] =='Other / Non-Disclosed').sum()
#other_total

purchase_value_total = gender_group["Price"].sum()
#purchase_value_total

average_purchase_price = gender_group["Price"].mean()
#average_purchase_price

average_per_person = gender_group["Price"].median()
#average_per_person

gender_purchase_df = pd.DataFrame(purchase_data_df["Gender"].value_counts())
gender_purchase_df["Purchase Count"] = (male_total, female_total, other_total)
gender_purchase_df["Average Purchase Price"] = average_purchase_price
gender_purchase_df["Total Purchase Value"] = purchase_value_total
gender_purchase_df["Average Total Purchase Per Person"] = average_per_person
gender_purchase_df.drop('Gender', inplace=True, axis=1)
gender_purchase_df["Average Purchase Price"] = average_purchase_price.map("${:,.2f}".format)
gender_purchase_df["Total Purchase Value"] = purchase_value_total.map("${:,.2f}".format)
gender_purchase_df["Average Total Purchase Per Person"] = average_per_person.map("${:,.2f}".format)
gender_purchase_df
```
